```python
'''
!= diferente
== igual
<= menor ou igual
>= maior ou igual
> maior
< menor
'''

tipoEscola = input("Estuda em colégiio: \n [1]Publico \n [2]Particular")
mediaAluno = int(input("Qual a média do aluno?"))
freqAluno = int(input("Qual a frequencia do aluno? (de 0 a 100)"))
nomeAluno = input("Qual o nome do aluno?")

if tipoEscola == "2":
    print("--------COLÉGIO PARTICULAR-----------")
    if mediaAluno >= 7 and freqAluno >= 70:
        print("O aluno " +str(nomeAluno)+ " foi aprovado com média " +str(mediaAluno))
    else:
        print("Reprovado")

if tipoEscola == "1":
    print("---------COLÉGIO PÚBLICO-----------")
    if mediaAluno >= 7 or freqAluno >= 70:
        print("O aluno " +str(nomeAluno)+ " foi aprovado com média " +str(mediaAluno))
    else:
        print("Reprovado")
```
